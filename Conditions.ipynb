```python
X,Y = input ("Enter number :").split()
for i in range(int(X),int(Y)):
    if i % 10 !=2 and i % 10 !=4:
        print(i, end =" ")
```

```python
N,X,Y = input ("Enter number :").split()
count = 0
for i in range(int(N),-200,-1):
    if i % int(X) == 0 and count<int(Y) and i>0:
        count+=1
        print(i, end =" ")
```

```python
N,X,Y = input ("Enter number :").split()
if N.isupper() or X.isupper() or Y.isupper():
    print("Yes")
else:
    print("No")
```

```python
N,X,Y = input ("Enter number :").split()
N = int(N)
X = int(X)
Y = int(Y)
if (N>0 and X>0 and Y>0) or (N<0 and X<0 and Y<0):
    print(N+X+Y)
elif (N<0 and X<0):
    print(N*X)
elif (X<0 and Y<0):
    print(Y*X)
elif (Y<0 and N<0):
    print(N*Y)
elif N<0:
    print(X+Y)
elif X<0:
    print(N+Y)
else:
    print(X+N)
```

```python
N = int(input("Enter number :"))
if N in range(1,11):
    print("Short")
elif N in range(11,51):
    print("Medium")
elif N in range(51,101):
    print("Long")
elif N > 100:
    print("Very Long")
```
